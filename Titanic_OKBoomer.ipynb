```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
data = pd.read_csv("https://raw.githubusercontent.com/plotly/datasets/master/titanic.csv")
data
```

```
generation = []
for age in data["Age"]:
  if age < 20:
    generation.append("Kid")
  elif age in range(21,40):
    generation.append("Millenial")
  elif age >=40:
    generation.append("OK Boomer")
  else:
    generation.append(np.nan)
generation
```

```
data["Generation"] = generation
data
```

```
data["Generation"].fillna("No Age", inplace = True)
data
```

```
data.groupby(data["Generation"]).count().plot.pie(y="Age")
```
